```python
df["sex"]=df["sex"].fillna(df["sex"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
plt.bar(df["sex"],df["body_mass_g"],color="skyblue")
plt.xlabel("flipper_length_mm")
plt.ylabel("body_mass_g")
plt.grid(alpha=0.5)
plt.show()
```

```python
df.head()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=df,  x="flipper_length_mm",y="body_mass_g",hue="sex")
sns.despine()
```

```python
df.tail()
```

```python
island_counts = df['island'].value_counts()
print(island_counts)

plt.figure(figsize=(7, 7))

plt.pie(
    island_counts,
    labels=island_counts.index,
    colors=['#facc15', '#38bdf8', '#f472b6'],
    autopct='%1.1f%%',
    explode=(0.05, 0.05, 0.05),
    shadow=True,
    startangle=90
)

plt.title('Penguin Island Distribution', fontsize=14)
plt.show()
```

```python
count_values = df["species"].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(
    count_values.values,
    labels=count_values.index,
    colors=['skyblue', 'yellow', 'purple'],
    autopct='%1.1f%%',
    explode=(0.05, 0.05, 0.05),
    shadow=True,
    startangle=90
)
plt.title('Penguin Species Distribution', fontsize=14)
plt.show()
```
